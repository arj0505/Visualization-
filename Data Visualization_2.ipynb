```python
df=pd.DataFrame(np.random.randn(100,3),columns=['A','B','C'])
#df.head()
df['A']=df['A'].cumsum() + 20
df['B']=df['B'].cumsum() + 20
df['C']=df['C'].cumsum() + 20
df.head()
```

```python
df.iplot() # be default kind='scatter'
```

```python
plt.plot(df)  #differance b/t above and below graph is just we need to 'i' before the plot. rest of the operations all are same
```

```python
df.plot()
```

```python
df.iplot(x='A',y='B',size=25,mode='markers')
```

```python
import seaborn as sns
titanic =sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.iplot(kind='bar',x='sex',y='survived',title='Survived',xTitle='Sex',yTitle='#Survived')
```

```python
titanic['sex'].value_counts() # value_counts() is group by count cloumns
```

```python
cf.getThemes() # theams means back ground 
               # cf is cufflinks. it is used configuration of plotly
```

```python
cf.set_config_file(theme='polar') # Here theam means back ground 
df.iplot(kind='bar',barmode='stack',bargap=0.5)
```

```python
cf.set_config_file(theme='henanigans')
df.iplot(kind='spread',barmode='stack',bargap=0.5)
```

```python
df.iplot(kind='box',barmode='stack',bargap=0.5)
```

```python
df.iplot(kind='barh',barmode='stack',bargap=0.5)
```

```python
df.iplot()  # By default is line graph plot
```

```python
df.iplot(kind='area',fill=True) #fill=True it will fill the area of gap
```

```python
help(cf.datagen)
```

```python
cf.datagen.sinwave(10,0.25) #datagen is generated with a random numbers
```

```python
cf.datagen.sinwave(10,0.25).iplot(kind='surface')  #it is a sinosudal wave graph
```

```python
cf.datagen.scatter3d(2,150,mode='stocks').iplot(kind='scatter3d',x='x',y='y',z='z')  #its scatter 3d graph
```

```python
df[['A','B','C']].iplot(kind='spread')
```

```python
df.iplot(kind='hist',bins=25,barmode='overlay',bargap=0.5)
```

```python
cf.datagen.bubble3d(5,4,mode='stocks').iplot(kind='bubble3d',x='x',y='y',z='z',size='size')
```

```python
cf.datagen.heatmap(20,20)
```

```python
cf.datagen.heatmap(20,20).iplot(kind = 'heatmap',colorscale='spectral',title='Cufflinks - Heatmap')
```
