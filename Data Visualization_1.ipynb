```python
import seaborn as sns
df = sns.load_dataset('iris')
print(type(df))
df.head()
```

```python
ax=df.plot(figsize=(10,6),title='Iris Dataset') #By default it will give line graph in matplotlib
ax.set_label('X_Axis')
ax.set_label('Y_Axis')
```

```python
ax1=df.plot(x = 'sepal_length', y = 'petal_length',kind='scatter',figsize=(10,5),title='Dataset')
ax1.set_xlabel('Sepal length values')
ax1.set_ylabel('Petal length values')
```

```python
ax1=df.plot(x = 'sepal_length', y = 'petal_length',kind='hist',figsize=(10,5),title='Dataset')
ax1.set_xlabel('Sepal length values')
ax1.set_ylabel('Petal length values')
```

```python
ax2=df.plot(x='sepal_width',y='petal_width',kind='bar',figsize=(18,5))
ax2.set_xlabel('sepal_width')
ax2.set_ylabel('petal_width')
```

```python
import seaborn as sns
df = sns.load_dataset('iris')
df.head()
```

```python
df=df.drop(['species'],axis=1)
df.head()
```

```python
df.diff().head()
```

```python
df.diff().plot(kind='hist',alpha=0.60,figsize=(4,3)) # alpha is used to adjust the transparency of a graph plot using the alpha attribute
```

By default, alpha=1. If you want to make the graph plot more transparent, then you can make alpha less than 1, such as 0.5 or 0.25.

```python
df.diff().hist(stacked=True,color='b',alpha=0.1,figsize=(10,10))
```

```python
df.plot.scatter(x='sepal_length',y='petal_length')
```

```python
#we can also represents same graph as  
df.plot(kind='scatter',x='sepal_length',y='petal_length')
```

```python
#if you want to represents 3 columns at a time
df.plot(kind='scatter',x='sepal_length',y='petal_length',c='sepal_width')  #parameter 'c' represents for 3rd axis
```

```python
df.head()
```

```python
df.plot.scatter(x='sepal_length',y='petal_length',c='sepal_width',s=100)
```

```python
df.plot.hexbin(x='sepal_length', y='petal_length', gridsize=5, C='sepal_width')
```

```python
#combine two columns graphs
ax=df.plot.scatter(x = 'sepal_length', y = 'petal_length', label = 'Length');
ax1=df.plot.scatter(x = 'sepal_width', y = 'petal_width', label = 'Width', ax = ax, color = 'r')
df.plot.scatter(x = 'sepal_width', y = 'petal_length', label = 'height', ax = ax1, color = 'g')
```
